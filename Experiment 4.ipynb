```python
import numpy as np
# Input (AND gate)
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

# Initialize
weights = np.zeros(2)
bias = 0
lr = 0.1
epochs = 10

# Training
for _ in range(epochs):
    for i in range(len(X)):
        z = np.dot(X[i], weights) + bias
        pred = 1 if z >= 0 else 0
        error = y[i] - pred

        weights += lr * error * X[i]
        bias += lr * error

# Output
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    pred = 1 if z >= 0 else 0
    print(X[i], "->", pred)
```
