```python
#Least Squares Regression using Formula Method
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

n = len(x)

# Required sums
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x * x)

# Least squares formula
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
c = (sum_y - m * sum_x) / n

print("Slope (m):", m)
print("Intercept (c):", c)

# Regression line
x_line = np.linspace(min(x), max(x), 100)
y_line = m * x_line + c

plt.scatter(x, y, label="Data Points")
plt.plot(x_line, y_line,color='black', label="Regression Line")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Least Squares Regression")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Logistic Regression using Formula Method

import numpy as np
import math
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + math.exp(-z))

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([0, 0, 1, 1, 1])

# Initialize parameters
b0 = 0.0
b1 = 0.0
alpha = 0.1
epochs = 1000

# Training using gradient descent
for _ in range(epochs):
    for i in range(len(x)):
        z = b0 + b1 * x[i]
        y_pred = sigmoid(z)
        error = y[i] - y_pred
        b0 += alpha * error
        b1 += alpha * error * x[i]

print("Bias:", b0)
print("Weight:", b1)

# Logistic curve
x_curve = np.linspace(min(x), max(x), 100)
y_curve = [sigmoid(b0 + b1 * xi) for xi in x_curve]

plt.scatter(x, y, label="Data Points")
plt.plot(x_curve, y_curve, label="Sigmoid Curve")
plt.xlabel("X values")
plt.ylabel("Probability")
plt.title("Logistic Regression")
plt.legend()
plt.grid(True)
plt.show()
```
